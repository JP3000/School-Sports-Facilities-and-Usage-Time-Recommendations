- pandas  
  - To clean and analyze data.  
  - Flatten JSON (json_normalize), read/write CSV, filter by placeCode (HBC/HBH/HSQ), concatenate DataFrames, select columns, handle missing values, save results.


```python
import pandas as pd

df = pd.read_csv('../data/processed_sports_facilities_11.csv')
df.head()
```

```python
# Filter out rows whose placeCode contains "HBC"
hbc_df = df[df['placeCode'].str.contains("HBC", na=False)]

# Filter out rows whose placeCode contains "HBH"
hbh_df = df[df['placeCode'].str.contains("HBH", na=False)]
```

```python
hbc_df.tail()
```

```python
# Concatenate hbc_df and hbh_df
badminton_df = pd.concat([hbc_df, hbh_df])
badminton_df.head()
```

```python
badminton_df['placeCode'].unique()
```

```python
badminton_df.to_csv('../data/badminton_df.csv', index=False)
```

```python
hsq_df = df[df['placeCode'].str.contains("HSQ", na=False)]
hsq_df.head()
```

```python
hsq_df.to_csv('../data/hsq_df.csv', index=False)
```
